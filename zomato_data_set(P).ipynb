```python
#approx cost should be a numeric column since it contains int values,
# though they are currently strings and contain commas for thousands and higher.
# We should clean this data and convert it to be able to properly analyze it.Â¶
```

```python
#remove commas
df['approx_cost_clean'] = df['approx_cost(for two people)'].str.replace(',', '')
#convert to numbers, coerce replaces any invalid parsing with NaN
df['approx_cost_clean'] = pd.to_numeric(df['approx_cost_clean'], errors='coerce')
```

```python
missing_values = df['approx_cost_clean'].isnull().sum()
missing_values
```

```python
df['approx_cost_clean'] = df['approx_cost_clean'].fillna(df['approx_cost_clean'].median())
```

top is new, but it would have to go if we want to transform the column into a numerical one to clean it. also NEW doesnt have any meaning in terms of a value for rate (number), so we cant replace it with some number to correspond to that. other problems include the fact that the values are rated out of 5, we need to rescale this. Also there are repeating values but the only difference is the spacing because the representation is string, this also should be cleaned.Â¶

```python
#clean the data
df['rate'] = df['rate'].replace(['NEW', '-', ' '], np.nan)
df['rate_clean'] = df['rate'].str.extract(r'(\d+\.\d+)', expand=False).astype(float)
```

```python
missing_values = df['rate_clean'].isnull().sum()
print("num of missing values in rate after cleaning: ", missing_values)
```
